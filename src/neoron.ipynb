## Manual Method of neuron output

```python
# inputs array consist of 4 inputs neorons
# there are three batches of data represented as rows in the array
x=np.array([[1,2,3,2.5],
            [2,5,-1,2],
            [-1.5,2.7,3.3,-.8]])

# weight of layer 1
w1=np.array([[.2, .8, -.5, 1],
            [.5,-.91, .26, -.5],
            [-.26, -.27, .17, .87]])

# weight of layer 2
w2=np.array([[.1, -.14, .5],
            [-.5,.12, -.33],
            [-.44, .73, -.13]])

# bias of layer 1
b1=np.array([2,3,.5])

# bias of layer 2
b2=np.array([-1,2,-.5])

layer1_output=np.dot(x,w1.T)+b1
layer2_output=np.dot(layer1_output,w2.T)+b2
print(layer2_output)
```

## Class method for neuron output

The dense_layer class create layer of neurons given the number of inputs to the neourns layer and the number of neurons in that layer.
The weights of the neurons is randomly generated
The output of the neurons layer is calculated through the function **forward()** given the input values.

```python
class dense_layer:
    # layer constructor
    def __init__(self,n_inputs,n_neurons):
        self.weights=0.01*np.random.randn(n_inputs,n_neurons)
        self.biases = np.zeros((1,n_neurons))
    # forward pass
    def forward(self,inputs):
        self.output=np.dot(inputs,self.weights)+self.biases

# create dataset
# y represent the category of points (0,1,2)
# x represent the points
x,y = spiral_data(samples=100,classes=3)

#create input layer
layer1=dense_layer(2,3)
layer1.forward(x)
print(layer1.output[:5])
        
```
